```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from neuralprophet import NeuralProphet
```

```python
import warnings

# Ignore all warnings
warnings.filterwarnings('ignore')
```

```python
# Load the dataset from the CSV file using pandas
df = pd.read_csv("https://github.com/ourownstory/neuralprophet-data/raw/main/kaggle-energy/datasets/tutorial01.csv")
```

```python
print(df.head(20))
```

```python
# Preprocessing on df['ds']
df['ds'] = pd.DatetimeIndex(df['ds'])
df.info()
```

```python
df['y'].plot()
```

```python
model = NeuralProphet(
    n_changepoints=10,
    # Enable all seasonality components
    yearly_seasonality=True,
    weekly_seasonality=True,
    daily_seasonality=True
    )
df_train, df_val = model.split_df(df, freq='D', valid_p = 0.2)
model.set_plotting_backend("plotly-static")
metrics = model.fit(df_train, freq='D', validation_df=df_val)
```

```python
# using last epoch results as training metrics summary
print(metrics[-1:])
```

```python
# forcast on next 30 days
future = model.make_future_dataframe(df, periods=30, n_historic_predictions=len(df))
forecast = model.predict(future)
```

```python
print(forecast['yhat1'][-30:])
```

```python
model.plot(forecast)
```

```python

df_residuals = pd.DataFrame({"ds": df["ds"], "residuals": df["y"] - forecast["yhat1"]})
fig = df_residuals.plot(x="ds", y="residuals", figsize=(10, 6))
```

```python
model.plot_components(forecast)
```

```python
model.plot_parameters(components=["seasonality"])
```
